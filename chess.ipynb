```python
import matplotlib.pyplot as plt

def draw_custom_chessboard():
    fig, ax = plt.subplots()

    # Create a chessboard with white and pink colors
    for i in range(8):
        for j in range(8):
            color = '#FFB6C1' if (i + j) % 2 == 0 else 'white'  # Pink and White colors
            ax.add_patch(plt.Rectangle((i, j), 1, 1, fill=True, color=color))

            # Add pieces on the board
            pieces = {
                (0, 0): '\u2656',  # White Rook
                (1, 0): '\u2658',  # White Knight
                (2, 0): '\u2657',  # White Bishop
                (3, 0): '\u2655',  # White Queen
                (4, 0): '\u2654',  # White King
                (5, 0): '\u2657',  # White Bishop
                (6, 0): '\u2658',  # White Knight
                (7, 0): '\u2656',  # White Rook
                (i, 1): '\u2659',  # White Pawn
                (0, 7): '\u265C',  # Black Rook
                (1, 7): '\u265E',  # Black Knight
                (2, 7): '\u265D',  # Black Bishop
                (3, 7): '\u265B',  # Black Queen
                (4, 7): '\u265A',  # Black King
                (5, 7): '\u265D',  # Black Bishop
                (6, 7): '\u265E',  # Black Knight
                (7, 7): '\u265C',  # Black Rook
                (i, 6): '\u265F',  # Black Pawn
            }

            if (i, j) in pieces:
                piece = pieces[(i, j)]
                color = 'black' if piece.islower() else 'white'
                ax.text(i + 0.5, j + 0.5, piece, fontsize=20, color=color, ha='center', va='center')

    # Add numbers 1-8 and letters a-h
    for i in range(8):
        ax.text(i + 0.5, -0.5, str(i + 1), fontsize=12, ha='center', va='center')
        ax.text(-0.5, i + 0.5, chr(ord('a') + i), fontsize=12, ha='center', va='center')

    # Set axis limits and remove ticks
    ax.set_xlim(-1, 8)
    ax.set_ylim(-1, 8)
    ax.set_xticks([])
    ax.set_yticks([])

    plt.show()

# Call the function to draw the custom chessboard with pieces, numbers, and letters
draw_custom_chessboard()
```
